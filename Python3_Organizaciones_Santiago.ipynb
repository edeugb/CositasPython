```python
sinNone = []
for i in tablaDatos:
    lugar = (str(i[1]) + ' SANTIAGO, CHILE')
    print(lugar)
    latlong = geocoder.komoot(lugar).latlng
# geocoder.google de repente dejó de funcionar y arrojaba puros None, por eso se cambia a komoot, osm, u otro
    print(latlong)
# También está geocoder.here(lugar) o geocoder.yahoo(lugar)
# https://geocoder.readthedocs.io/providers/HERE.html#geocoding
# https://geocoder.readthedocs.io/providers/OpenStreetMap.html
    nombre = i[0]
#     print(nombre)
    if latlong != None:
        sinNone.append([nombre,latlong])
sinNone
```

```python
mapa = folium.Map(location=[-33.437886, -70.650661], zoom_start=10)#,"Stamen Terrain"
for i in sinNone:
    popup = folium.Popup(i[0], parse_html=True)
#     print(i[1])
#     print(i[0])
    folium.Marker(location=i[1],popup=popup).add_to(mapa)
mapa

#Había un problema con el popup del Marker

# La solución que adapté fue esta
# for point in range(len(locationlist)):
#     popup = folium.Popup(labels[point], parse_html=True)
#     folium.Marker(locationlist[point], popup=popup).add_to(m)

# Pero hay ptra que debería servir adaptando lo siguiente
# import html
# for nombre, x, y in tablaDatos:
#     ubicacion = [float(y),float(x)]
#     nombre = html.escape(nombre)
#     folium.Marker(ubicacion, popup=nombre).add_to(mapaCiclovias)
    
```
