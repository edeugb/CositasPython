```python
from robobrowser import RoboBrowser
import folium
import geocoder
import pandas
```

```python
browser = RoboBrowser(parser="html5lib")
url = "http://www.t13.cl/noticia/nacional/el-ranking-colegios-mejor-puntaje-promedio-psu-2017"
browser.open(url)
browser
```

```python
browser.select('#main-article table') #lo de main article esta en el script de la página
```

```python
tablaDatos = browser.select('#main-article table')[0] #Lo sacas de la lista
```

```python
datos = tablaDatos.__str__() #Lo transformo a string para pandas
datos[:1000]
```

```python
df = pandas.read_html(datos, header = 0)[0] #primer argumento de read_html recepciona string, por eso lo anterior
df.head() #Hay que cachar cómo se visualizan los datos. Ojo en el promedio. En este caso hay que separar por "." los decimales que estaban por ","
```

```python
datos = datos.replace(",",".")
datos = datos.replace("Ñ","N")
datos = datos.replace("Ó","O")
datos = datos.replace("´","'")
datos = datos.replace("'","'")
datos[:1000] #Ahora está el string listo
```

```python
df = pandas.read_html(datos, header = 0)[0] #primer argumento de read_html recepciona string, por eso lo anterior
df.head()
# df['NOMBRE COLEGIO'].str.replace("Ñ","N") SI NO LO HUBIESE HECHO EN DATOS SE VERÍA ASÍ DESPUÉS PARA EL DATA FRAME
df['COMUNA'].str.replace("Ó","O")
```

```python
# Ahora hay que ubicar todos los puntos, para lo que necesito nombre del colegio, ubicación 
# y el promedio del puntaje que será proporcional al tamañano del círculo *0,05
# También hay que limpiar los Nones

# sinNone = []
# for i in tablaDatos:
#     lugar = (i[0]+', '+i[1]+', CHILE')
#     latlong = geocoder.google(lugar).latlng
#     if latlong != None:
#         sinNone.append(latlong)
# sinNone

# Sin embargo, lo de arriba si bien funciona, no es factible porque pierde la referencia 
# con las otras columnas, por lo que hay que eliminar el elemnto entero
```

```python
# LIMPIAR UBICACIONES ERRONEAS (NONE)
# listaUbicaciones=[[nombre, tipo, capitalLocal, latitudLongitud] for nombre, tipo, capitalLocal, latitudLongitud in listaUbicaciones if latitudLongitud != None]
sinNone = []
for i in tablaDatos:
    lugar = (i[0]+', '+i[1]+', CHILE')
    latlong = geocoder.google(lugar).latlng
# También está geocoder.here(lugar) o geocoder.yahoo(lugar)
# https://geocoder.readthedocs.io/providers/HERE.html#geocoding
# https://geocoder.readthedocs.io/providers/OpenStreetMap.html
    nombre = i[0]
    prom = i[3]
    if latlong != None:
        sinNone.append([nombre,prom,latlong])
sinNone
```

```python
mapa2 = folium.Map(location=[-33.438182, -70.5685989],tiles="Stamen Toner")#"Stamen Terrain"
for i in sinNone:
    folium.CircleMarker(location=i[2],
                        radius=i[1]*0.005,
                        popup=i[0],
                        color='black',
                        fill=True,
                        fill_color='black',
                    fill_opacity=1).add_to(mapa2)
mapa2
# mapa2.save("mapa.html")
#luego abro mapa que ya se guardó como html
```
